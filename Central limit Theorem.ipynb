```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
```

```python
df = pd.DataFrame(np.random.randint(0, 1000, size=(100000,)))
df.describe()
```

```python
sampleSize = 30
def distribution(sampleSize, n=100):
    data = np.array([int(df.sample(sampleSize).mean())
            for __ in range(0,n)])
    sns.kdeplot(data, shade=True, color="g")
    title = "Distribution for sample size = {}, Obtained mean = {}"
    plt.title(title.format(sampleSize, data.mean()))
    plt.xlabel("Mean of Random Sample")
    plt.show()

```

```python
for sampleSize in [2, 100, 500, 1000, 5000, 10000]:
    distribution(sampleSize)
```
